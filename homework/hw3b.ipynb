## Paper: The Foundations of Cost-Sensitive Learning

### Background

Elkan's "The Foundations of Cost-Sensitive Learning" (2001) is a short, influential paper (~8 pages) that argues for a specific approach to handling unequal misclassification costs. Rather than manipulating the training data (e.g., oversampling or undersampling), Elkan argues that the right approach is to learn good probability estimates and then apply costs at *decision time* by adjusting the classification threshold.

This paper directly connects to what you'll do in HW3c: instead of resampling data, you'll tune thresholds and apply class weights to optimize for business-relevant costs.

Read the paper (available on Canvas). Focus on:

- Sections 1-3: the core argument for cost-sensitive decision-making
- Section 4: the relationship between costs and optimal thresholds
- Section 6: the comparison with resampling approaches

You can skim or skip: Section 5 (mathematical proofs) and Section 7 (multi-class extensions).
